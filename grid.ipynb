Q3.Explain the use of grid-auto-row and grid-auto-column using code examples.

The grid-auto-rows and grid-auto-columns properties in CSS Grid are used to define the size of implicitly created rows or columnsâ€”i.e., those that are not explicitly defined in the grid-template-rows or grid-template-columns.

1.grid-auto-rows
Purpose: Specifies the height of automatically created rows.

Used when: Grid items are placed in rows that aren't explicitly defined.

```python
<style>
  .grid-container {
    display: grid;
    grid-template-columns: 100px 100px;
    grid-auto-rows: 80px;
    gap: 10px;
    background: #f3f3f3;
  }

  .grid-item {
    background: lightblue;
    padding: 10px;
    text-align: center;
  }
</style>

<div class="grid-container">
  <div class="grid-item">1</div>
  <div class="grid-item">2</div>
  <div class="grid-item">3</div> <!-- Implicit row -->
  <div class="grid-item">4</div>
</div>
```

2.grid-auto-columns
Purpose: Specifies the width of automatically created columns.

Used when: Grid items are placed in columns that aren't explicitly defined.

```python
<style>
  .grid-container {
    display: grid;
    grid-template-rows: 100px;
    grid-auto-columns: 120px;
    grid-auto-flow: column;
    gap: 10px;
    background: #fafafa;
  }

  .grid-item {
    background: lightgreen;
    padding: 10px;
    text-align: center;
  }
</style>

<div class="grid-container">
  <div class="grid-item">A</div>
  <div class="grid-item">B</div>
  <div class="grid-item">C</div> <!-- Implicit column -->
</div>
```

Q5.Explain the difference between justify-items and justify-self using code examples.

Both justify-items and justify-self play important roles in controlling the horizontal alignment of elements within CSS Grid layouts, but they operate at different levels. The justify-items property is set on the grid container and establishes the default horizontal alignment for all child grid items. It accepts values such as start, center, end, and stretch, which determine how each item is positioned within its grid cell along the inline (horizontal) axis. For example, justify-items: center centers every grid item horizontally in their respective cells, providing a uniform alignment across the entire grid.

In contrast, justify-self targets individual grid items and overrides the containerâ€™s justify-items value for that specific item only. This makes it useful when you want most items to follow a general alignment rule but need one or a few items to behave differently. For example, if the containerâ€™s justify-items is set to start, all items will align to the left, but an item with justify-self: end will align to the right inside its grid cell. This granular control allows developers to create flexible and responsive layouts without having to restructure the entire grid.

Both properties enhance the versatility of CSS Grid by separating concerns between global container-level alignment (justify-items) and local item-level overrides (justify-self). They complement each other to help designers and developers create visually balanced and precise layouts. Additionally, these properties have vertical counterparts, align-items and align-self, which work along the block (vertical) axis in a similar way. Understanding and using justify-items and justify-self effectively allows for fine control over the positioning of elements, improving both aesthetics and usability in modern web design.
